```python
from transformers import BlipProcessor, BlipForConditionalGeneration
import torch
from PIL import Image
import random
import gradio as gr

# Load BLIP Processor & Model
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-large")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-large")

# Move model to GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

def generate_instagram_caption(image_caption):
    """
    Generate an engaging Instagram caption based on the image caption.
    """
    insta_phrases = [
        "A picture is worth a thousand words, but this one is priceless! ",
        "Captured a moment, now sharing the story! ",
        "Some moments deserve a permanent place on my feed! ",
        "Bringing vibes to your feed! ",
        "What a day! Just had to share this one! "
    ]
    return random.choice(insta_phrases) + image_caption

def generate_hashtags(image_caption):
    """
    Generate relevant hashtags based on the image caption.
    """
    words = image_caption.lower().split()
    hashtags = [f"#{word}" for word in words if len(word) > 3]  # Convert words to hashtags
    common_tags = ["#InstaGood", "#PhotoOfTheDay", "#Beautiful", "#Love", "#Nature"]

    # Select up to 3 common tags and 5 AI-generated hashtags
    return " ".join(random.sample(common_tags, min(3, len(common_tags))) + hashtags[:5])

def generate_caption(image_path):
    """
    Given an image, generate captions for general use and Instagram.
    """
    try:
        # Open and preprocess image
        image = Image.open(image_path).convert("RGB")
        inputs = processor(image, return_tensors="pt").to(device)

        # Generate caption
        with torch.no_grad():
            caption_ids = model.generate(**inputs)

        image_caption = processor.decode(caption_ids[0], skip_special_tokens=True)
        insta_caption = generate_instagram_caption(image_caption)
        hashtags = generate_hashtags(image_caption)

        return image_caption, insta_caption, hashtags

    except Exception as e:
        return f"❌ Error: {str(e)}", "", ""

# Create Gradio Interface
interface = gr.Interface(
    fn=generate_caption,  # Function to call
    inputs=gr.Image(type="filepath"),  # Image input
    outputs=[
        gr.Textbox(label="Generated Caption"),
        gr.Textbox(label="Instagram Caption"),
        gr.Textbox(label="Hashtags")
    ],
    title="📷 AI Image Captioning",
    description="Upload an image and get a caption, an Instagram-friendly caption, and relevant hashtags using BLIP.",
)

# Launch Gradio App
interface.launch()
```
